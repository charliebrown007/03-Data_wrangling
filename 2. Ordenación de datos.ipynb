#### <font color=#006700> Uso de la función sorted para ordenar una columna

Sort by the values along either axis.<br/>
PARAMETERS:<br/>
**by**: str or list of str
Name or list of names to sort by.
- if axis is 0 or ‘index’ then by may contain index levels and/or column labels.
- if axis is 1 or ‘columns’ then by may contain column levels and/or index labels.

**axis**: {0 or ‘index’, 1 or ‘columns’}, default 0.
- Axis to be sorted.

**ascending**: bool or list of bool, default True
- Sort ascending vs. descending. Specify list for multiple sort orders. If this is a list of bools, must match the length of the by.

**inplace**: bool, default False
- If True, perform operation in-place.

**kind**: {‘quicksort’, ‘mergesort’, ‘heapsort’, ‘stable’}, default ‘quicksort’<br/>
- Choice of sorting algorithm. See also numpy.sort() for more information. mergesort and stable are the only stable algorithms. For DataFrames, this option is only applied when sorting on a single column or label.

**na_position**: {‘first’, ‘last’}, default ‘last’<br/>
- Puts NaNs at the beginning if first; last puts NaNs at the end.

**ignore_index**: bool, default False<br/>
- If True, the resulting axis will be labeled 0, 1, …, n - 1.

```python
# Orden ascendente
df1 = df.sort_values(by = ['air_time'], inplace = False)
df1.head()
```

```python
# Orden descendente
df1 = df.sort_values(by = ['air_time'], inplace = False, ascending = False)
df1.head()
```

#### <font color=#006700> Ordenar valores según múltiples columnas

Considerar que el orden en el que se describan las columnas dentro del argumento 'by' determinarán la forma de romper el empate en la operación.

```python
# Ordena por año, mes, día y hora de salida
df1 = df.sort_values(by = ['year','month','day','dep_time'], inplace = False, ascending = True)
df1.head()
```
