## Export Data

### Librerías

```python
import pandas as pd
import seaborn as sns
```

### Carga de datos

```python
data = sns.load_dataset('planets')
data.head()
```

### Exportar datos

#### 1. Formato: **csv**

```python
# Registros con años pares
# Módulo: %
# data_01 = (data['year']%2) == 0 # Máscara
data_01 = data.loc[(data['year']%2) == 0].reset_index(drop = True)[:10]
data_01
```

```python
# Separadores
# Coma: ','
# Punto y coma: ';'
# Espacio: ' '
# Tabulador: '\t'


# Exportar tabla
data_01.to_csv(path_or_buf = 'data_01.csv', # Dirección para guardar el archivo o simplemente el nombre acompañado de .csv
               sep = ';',
               header = True, # Exportar encabezado del dataframe. Por defecto es True.
               index = False, # Exportar índices de fila. Por defecto es True.
               encoding = 'utf-8')  # Codificación. Por defecto es utf-8
```

#### 2. Formato: **txt**

El método es casi idéntico al anterior, solo se debe cuidar que el nombre del archivo tenga la extensión ".txt"

```python
# Registros con años impares
# Módulo: %
# data_01 = (data['year']%2) != 0 # Máscara
data_02 = data.loc[(data['year']%2) != 0].reset_index(drop = True)[:10]
data_02
```

```python
# Separadores
# Coma: ','
# Punto y coma: ';'
# Espacio: ' '
# Tabulador: '\t'


# Exportar tabla
data_02.to_csv(path_or_buf = 'data_02.txt', # Dirección para guardar el archivo o simplemente el nombre acompañado de .csv
               sep = '\t',
               header = True, # Exportar encabezado del dataframe. Por defecto es True.
               index = True, # Exportar índices de fila. Por defecto es True.
               encoding = 'utf-8')  # Codificación. Por defecto es utf-8
```

#### 3. Formato: **xlsx**

```python
# Planetas con masa superior al tercer cuartil
data_03 = data.loc[data['mass'] > data['mass'].quantile(q = 0.75)].reset_index(drop = True)[:10]
data_03
```

```python
data_03.to_excel(excel_writer = 'data_03.xlsx',
                 sheet_name = 'Dataset 03',
                 float_format = '%.2f',
                 index = False,
                 header = True,
                 startcol = 1, # Inicio de la escritura en una columna
                 startrow = 1) # Inicio de la escritura en una fila
```
